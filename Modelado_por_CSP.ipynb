Modelado por CSP :

Un conjunto de Variables,
un conjunto de Dominios (Un dominio asociado a cada variable) y un conjunto de restricciones.

Objetivo: Lograr que cada variable tenga un valor asociado (Encontrar una solución), garantizando que todos los valores de todas las variables con consistentes a todas las restricciones

Como lo hacemos?
CSP -> Tecnicas de consistencia -> CSP K - Consistente -> Tecnicas de Busqueda -> Solucion

1. Conjunto variables
2. Conjunto de dominios
3. Conjunto de restricciones (27 restricciones)

```python
# ══════════════════════════════════════════════════════
#  FUNCIONES DE PROPAGACIÓN
# ══════════════════════════════════════════════════════

def AllDif(VarsDoms, VarsList):
    for varSrc in VarsList:
        if len(VarsDoms[varSrc]) == 1:
            for varDst in VarsList:
                if varSrc != varDst:
                    VarsDoms[varDst] = VarsDoms[varDst] - VarsDoms[varSrc]

def ExcValue(VarsDoms, VarsList):
    for varSrc in VarsList:
        if len(VarsDoms[varSrc]) > 1:
            U = set()
            for varDst in VarsList:
                if varSrc != varDst:
                    if len(VarsDoms[varDst]) > 1:
                        U = U.union(VarsDoms[varDst])
            Dif = VarsDoms[varSrc] - U
            if len(Dif) == 1:
                VarsDoms[varSrc] = Dif

def propagar(VarsDoms, ConstraintsVarsLists):
    cambio = True
    while cambio:
        anterior = {v: frozenset(d) for v, d in VarsDoms.items()}
        for constraint in ConstraintsVarsLists:
            AllDif(VarsDoms, constraint)
            ExcValue(VarsDoms, constraint)
        cambio = any(VarsDoms[v] != anterior[v] for v in VarsDoms)

print("✅ Funciones definidas correctamente")
```

```python
# ══════════════════════════════════════════════════════
#  RESOLVER Y DIAGNOSTICAR
# ══════════════════════════════════════════════════════

propagar(VarsDoms, ConstraintsVarsLists)

resueltas    = sum(1 for d in VarsDoms.values() if len(d) == 1)
sin_resolver = sum(1 for d in VarsDoms.values() if len(d) > 1)

print(f"Celdas resueltas:    {resueltas} / 81")
print(f"Celdas sin resolver: {sin_resolver}")

if sin_resolver == 0:
    print("\n✅ ¡Sudoku resuelto solo con propagación!")
else:
    print("\n⚠️  Se necesita backtracking.")
```

```python
# ══════════════════════════════════════════════════════
#  BACKTRACKING
# ══════════════════════════════════════════════════════

import copy

def es_valido(VarsDoms):
    """Verifica que ningún dominio esté vacío."""
    return all(len(d) > 0 for d in VarsDoms.values())

def esta_resuelto(VarsDoms):
    """Verifica que todas las variables tengan exactamente un valor."""
    return all(len(d) == 1 for d in VarsDoms.values())

def elegir_variable(VarsDoms):
    """
    Heurística MRV (Minimum Remaining Values):
    Elige la variable no resuelta con el dominio más pequeño.
    Esto reduce el árbol de búsqueda significativamente.
    """
    return min(
        (v for v in VarsDoms if len(VarsDoms[v]) > 1),
        key=lambda v: len(VarsDoms[v])
    )

def backtracking(VarsDoms, ConstraintsVarsLists):
    """
    Algoritmo principal de búsqueda con retroceso.
    En cada paso:
      1. Propaga restricciones
      2. Si está resuelto, retorna la solución
      3. Si hay contradicción, retorna None
      4. Si no, elige una variable, prueba cada valor
         y recurre
    """
    # Propagar antes de buscar
    propagar(VarsDoms, ConstraintsVarsLists)

    if not es_valido(VarsDoms):
        return None                          # Contradicción

    if esta_resuelto(VarsDoms):
        return VarsDoms                      # ✅ Solución encontrada

    # Elegir la variable más restringida
    var = elegir_variable(VarsDoms)

    # Probar cada valor posible
    for valor in VarsDoms[var]:
        # Copia profunda para no contaminar el estado actual
        nuevo_estado = copy.deepcopy(VarsDoms)
        nuevo_estado[var] = {valor}          # Asignar el valor

        resultado = backtracking(nuevo_estado, ConstraintsVarsLists)
        if resultado is not None:
            return resultado                 # Solución encontrada

    return None                              # Ningún valor funcionó

print("✅ Backtracking definido correctamente")
```

```python
# ══════════════════════════════════════════════════════
#  RESOLVER CON BACKTRACKING
# ══════════════════════════════════════════════════════

solucion = backtracking(VarsDoms, ConstraintsVarsLists)

if solucion:
    resueltas = sum(1 for d in solucion.values() if len(d) == 1)
    print(f"✅ ¡Sudoku resuelto! Celdas resueltas: {resueltas} / 81")
else:
    print("❌ No se encontró solución.")
```
